```
from bs4 import BeautifulSoup
from bs4 import UnicodeDammit
import requests
from html.parser import HTMLParser
htmlparser = HTMLParser()
```

```
resp = requests.get('http://cavitexpressway.com/traffic-updates-feeds')
html = htmlparser.unescape(resp.text)
soup = BeautifulSoup(html, 'lxml')
```

```
test = soup.find('div', {'class': ['field-item even']})
print(test.get_text())
```
